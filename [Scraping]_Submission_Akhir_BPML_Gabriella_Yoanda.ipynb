# SCRAPING ULASAN APLIKASI TOKOPEDIA PADA GOOGLE PLAY

## Import Library

```python
!pip install google-play-scraper

from google_play_scraper import app, reviews, Sort

import pandas as pd
pd.options.mode.chained_assignment = None
import numpy as np
```

## Scraping Dataset

Digunakan ulasan aplikasi Tokopedia yang ada di Google Play sebanyak 15.000 ulasan terbaru yang penggunanya dari Indonesia dan menuliskan ulasan dengan Bahasa Indonesia.

```python
result, _ = reviews(
    'com.tokopedia.tkpd',
    lang='id',
    country='id',
    sort=Sort.NEWEST,
    count=15000
)
```

```python
df = pd.DataFrame(result)
df
```

Menyimpan ulasan dalam file CSV

```python
df.to_csv('ulasan_tokopedia.csv', index=False)
```
